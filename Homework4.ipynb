```python
n = int(input("Nhập n: "))
chan = 0
le = 0

while n > 0:
    digit = n % 10
    if digit % 2 == 0:
        chan = chan + 1
    else:
        le = le + 1
    n = n // 10

print("Số chữ số chẵn:", chan)
print("Số chữ số lẻ:", le)
```

### W4A25

```python
x = int(input())

if x == -1:
    print("Không có dữ liệu")
else:
    max = x
    min = x

    while x != -1:
        if x > max:
            max = x
        if x < min:
            min = x
        x = int(input("Nhập số -1 để dừng: "))

    print("Max =", max)
    print("Min =", min)
```

```python
A = int(input("Nhập A: "))

a = 1
b = 1

if A < 1:
    print("Không có")
else:
    while b <= A:
        temp = a + b
        a = b
        b = temp

    print("Số fibonacci lớn nhất <=", A, "là:", a)
```

### W4A27

```python
s = input("Nhập chuỗi: ")

dem = 0
trong_tu = 0   

i = 0
while True:
    try:
        c = s[i]
    except:
        break

    if c != ' ':
        if trong_tu == 0:
            dem = dem + 1
            trong_tu = 1
    else:
        trong_tu = 0

    i = i + 1

print("Số từ:", dem)
```

### W4A29

```python
s = input("Nhập chuỗi: ")
kq = ""
i = 0
while i < len(s):
    c = s[i]
    if 'a' <= c <= 'z':
        kq = kq + chr(ord(c) - 32)
    elif 'A' <= c <= 'Z':
        kq = kq + chr(ord(c) + 32)
    else:
        kq = kq + c
    i = i + 1
print(kq)
```

### W4A30

```python
s = input("Nhập: ")
tong = 0
num = 0
dau = 1  
i = 0
while i < len(s):
    c = s[i]
    if c == '-':
        dau = -1
    elif c >= '0' and c <= '9':
        num = num * 10 + (ord(c) - ord('0'))
    else:
        tong = tong + dau * num
        num = 0
        dau = 1
    i = i + 1
tong = tong + dau * num  
print("Tong:", tong)
```

### W4A31

```python
s = input("Nhập chuỗi: ")
hoa = 0
thuong = 0
so = 0
i = 0
while i < len(s):
    c = s[i]
    if 'A' <= c <= 'Z':
        hoa = hoa + 1
    elif 'a' <= c <= 'z':
        thuong = thuong + 1
    elif '0' <= c <= '9':
        so = so + 1
    i = i + 1
print("Chữ hoa: ", hoa)
print("Chữ thường:", thuong)
print("Chữ số:", so)
```
