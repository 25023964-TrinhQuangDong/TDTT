### W4A6

```python
t = int(input("Nhập số bộ test: "))

while t > 0:
    a, b = map(int, input("Nhập a, b: ").split())
    total = 0

    i = a
    while i <= b:
        if i >= 2:
            snt = True
            j = 2
            while j * j <= i:
                if i % j == 0:
                    snt = False
                    break
                j += 1
            if snt:
                total += i
        i += 1

    print("Tổng:", total)
    t -= 1
```

### W4A8

```python
n = int(input("Nhập số: "))
steps = 0

while True:
    temp = n
    rev = 0

    while temp > 0:
        rev = rev * 10 + temp % 10
        temp //= 10

    if rev == n:
        break

    n = n + rev
    steps += 1

print("Số bước:", steps)
print("Palindrome:", n)
```

### W4A9

```python
n = int(input())
found = False
i = 1
while i * i <= n:
    x = i * i
    temp = x
    used = [0] * 10
    distinct = True
    if temp == 0:
        used[0] = 1
    while temp > 0:
        digit = temp % 10
        if used[digit] == 1:
            distinct = False
            break
        used[digit] = 1
        temp //= 10
    if distinct:
        print(x, end=" ")
        found = True
    i += 1
if not found:
    print("no number")
```

### W4A10

```python
n = int(input())

max_len = 0
result = 1

i = 1
while i <= n:
    x = i
    length = 1

    while x != 1:
        if x % 2 == 0:
            x //= 2
        else:
            x = 3 * x + 1
        length += 1

    if length > max_len:
        max_len = length
        result = i

    i += 1

print(result)
```

```python
a = int(input("Nhập số thứ nhất: "))
b = int(input("Nhập số thứ hai: "))
tong_a = 0
i = 1
while i < a:
    if a % i == 0:
        tong_a += i
    i += 1

tong_b = 0
i = 1
while i < b:
    if b % i == 0:
        tong_b += i
    i += 1

if tong_a == b and tong_b == a:
    print("true")
else:
    print("false")
```
