```python
def nguyen_to(n):
    if n<2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n%i == 0:
            return False
        return True
```

### W5A4 

```python
def so_hoan_hao(n):
    if n < 2:
        return False
    total = 1
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            total += i
            if i != n // i:
                total += n // i
    return total == n
```

```python
def vi_tri_dau_tien(lst, k):
    for i in range(len(lst)):
        if lst[i] == k:
            return i
    return -1
```

### W5A6

```python
def giai_thua(n):
    if n<0:
        return None
    elif n==0 or n==1:
        return 1
    else:
        ketqua = 1
        for i in range(2, n+1):
            ketqua *= i
            return ketqua
```

### W5A7

```python
def may_tinh(num1, num2, operator):
    if operator == '+':
        result = num1 + num2
    elif operator == '-':
        result = num1 - num2
    elif operator == '*':
        result = num1 * num2
    elif operator == '/':
        if num2 == 0:
            return "Không thể chia cho 0"
        result = num1 / num2
    else:
        return "Toán tử không hợp lệ"
    
    return round(result, 2)
```

### W5A8

```python
def hamming(x, y):
    xor = x ^ y
    count = 0
    while xor:
        count += xor & 1
        xor >>= 1
    return count
```

### W5A9

```python
def tong_cac_chu_so(n):
    total = 0
    while n > 0:
        total += n % 10
        n //= 10
    return total
```

### W5A10

```python
def dang_cau(a, b):
    if len(a) != len(b):
        return False
    map_a_to_b = {}
    map_b_to_a = {}
    for char_a, char_b in zip(a, b):
        if char_a in map_a_to_b:
            if map_a_to_b[char_a] != char_b:
                return False
        else:
            map_a_to_b[char_a] = char_b
        
        if char_b in map_b_to_a:
            if map_b_to_a[char_b] != char_a:
                return False
        else:
            map_b_to_a[char_b] = char_a
    
    return True
```

```python
def canh_tam_giac_vuong(a, b, c):
    tgv=False
    if a >= b and a >= c:
        ch = a
        x, y = b, c
    elif b >= a and b >= c:
        ch = b
        x, y = a, c
    else:
        ch = c
        x, y = a, b
    if ch**2 == x**2 + y**2:
        tgv = True
    return tgv
```
