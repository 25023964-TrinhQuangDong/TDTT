Máy tính giao tiếp với con người thông qua các thiết bị thu nhận các yêu cầu của người dùng
(input devices) như bàn phím, con chuột, microphone và các thiết bị hiển thị kết quả tính
toán (output devices) như màn hình, loa. Trong bài này, chúng ta sẽ làm việc với thiết bị
hiển thị kết quả tính toán đơn giản nhất, đó là màn hình văn bản (text screen). Màn hình văn
bản được dùng để hiển thị các thông tin dưới dạng văn bản từ chữ cái (characters), chữ số
(digits) đến các dòng văn bản (text line), các đoạn văn bản (text paragraph). Trong đa số
trường hợp, màn hình văn bản còn được gọi là thiết bị xuất chuẩn (standard output device)
bởi vì nó là thiết bị hiển thị và đầu ra mặc định của các chương trình.
Trong python hàm print() được dùng để hiển thị thông tin ra màn hình (console), nó là một
trong những công cụ quan trọng nhất để: Hiển thị kết quả cho người dùng, Kiểm tra, debug
chương trình khi lập trình, Xuất thông tin định dạng đẹp... Cú pháp cơ bản của hàm:
print(*objects, sep=' ', end='\n', file=sys.stdout, flush=False).
Các đối số hay dùng:
'- *objects: Chuỗi, số, danh sách, v.v... cần in ra.
'- sep: Dấu phân cách giữa các đối tượng (mặc định là khoảng trắng ' ').
'- end: Ký tự kết thúc dòng (mặc định là xuống dòng '\n').
'- flush: Ép ghi ra màn hình ngay lập tức.
print() là một trong những hàm được Python cung cấp sẵn, bạn có thể dùng nó mà không cần
import thư viện gì hết.
Nó được định nghĩa trong mô-đun builtins, là mô-đun mặc định luôn được nạp khi Python
chạy.
Như vậy, toàn bộ một chương trình in ra dòng chữ "Hello World!" như sau: print("Hello
World!").
Hãy viết chương trình in ra dòng chữ "Hello World!" vào ô dưới đây rồi ấn Nộp bài để kiểm
tra xem bạn đã biết cách in một thông báo ra màn hình.

```python
print("Hello World!")
```

[AverageOfSix]
Viết chương trình python để tính điểm trung bình môn học của một học sinh trong lớp. Có tổng cộng
6 đầu điểm: a1, b1, c1, a2, b2, a3. Trong đó:
• Các điểm a1, b1, c1 được nhân hệ số 1,
• Các điểm a2, b2 được nhân hệ số 2,
• Điểm a3 được nhân hệ số 3.
Công thức tính điểm trung bình là: TB=((a1+b1+c1)+(a2+b2)×2+a3×3)/10
Đầu vào: Trong bài tập này, dữ liệu vào chỉ gồm duy nhất một dòng bao gồm 6 số thể hiện cho 6
đầu điểm a1, b1, c1, a2, b2, a3 theo đúng thứ tự đó. Trong đó, hai đầu điểm liên tiếp nhau sẽ được
phân tách nhau bởi duy nhất 1 dấu cách.
Đầu ra: Đầu ra là giá trị điểm trung bình môn học của 6 môn học, lấy đến 1 chữ số phần thập phân.
