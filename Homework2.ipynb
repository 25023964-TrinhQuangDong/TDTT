### Bài 3

```python
a = float(input("Nhập cạnh a: "))
b = float(input("Nhập cạnh b: "))
c = float(input("Nhập cạnh c: "))
if a + b <= c or a + c <= b or b + c <= a:
    print("Không là ba cạnh của một tam giác")
else:
    if a == b == c:
        loai = "Tam giác đều"
    elif a == b or b == c or a == c:
        loai = "Tam giác cân"
    elif a*a + b*b == c*c or a*a + c*c == b*b or b*b + c*c == a*a:
        loai = "Tam giác vuông"
    else:
        loai = "Tam giác thường"
    p = a + b + c
    s = p / 2
    dien_tich = (s * (s - a) * (s - b) * (s - c))**0.5
    print("Loại:", loai)
    print("Chu vi:", p)
    print("Diện tích:", dien_tich)
```

```python
a = float(input("Nhập a: "))
b = float(input("Nhập b: "))
c = float(input("Nhập c: "))
if a == 0:
    if b == 0:
        if c == 0:
            print("Vô số nghiệm")
        else:
            print("Vô nghiệm")
    else:
        print("Nghiệm x =", -c / b)
else:
    delta = b*b - 4*a*c

    if delta < 0:
        print("Phương trình vô nghiệm")
    elif delta == 0:
        x = -b / (2*a)
        print("Nghiệm kép x =", x)
    else:
        sqrt_delta = (delta)**0.5
        x1 = (-b + sqrt_delta) / (2*a)
        x2 = (-b - sqrt_delta) / (2*a)
        print("x1 =", x1)
        print("x2 =", x2)
```

### Bài 8

```python
a = float(input("Nhập a: "))
b = float(input("Nhập b: "))
c = float(input("Nhập c: "))
d = float(input("Nhập d: "))
m = float(input("Nhập m: "))
n = float(input("Nhập n: "))
D = a*d - b*c
Dx = m*d - b*n
Dy = a*n - m*c
if D == 0:
    if Dx == 0 and Dy == 0:
        print("Hệ có vô số nghiệm")
    else:
        print("Hệ vô nghiệm")
else:
    x = Dx / D
    y = Dy / D
    print("x =", x)
    print("y =", y)
```

### Bài 10

```python
x0 = float(input("Nhập tâm x0: "))
y0 = float(input("Nhập tâm y0: "))
r = float(input("Nhập bán kính: "))
x = float(input("Nhập điểm x: "))
y = float(input("Nhập điểm y: "))
left = (x - x0)*(x - x0) + (y - y0)*(y - y0)
right = r * r
if left == right:
    print("Điểm nằm trên đường tròn")
elif left < right:
    print("Điểm nằm trong đường tròn")
else:
    print("Điểm nằm ngoài đường tròn")
```

### Bài 11

```python
x = float(input("Nhập x: "))
y = int(input("Nhập y (số nguyên): "))
ket_qua = 1
if y > 0:
    i = 0
    while i < y:
        ket_qua = ket_qua * x
        i = i + 1
elif y < 0:
    i = 0
    while i < -y:
        ket_qua = ket_qua * x
        i = i + 1
    ket_qua = 1 / ket_qua
else:
    ket_qua = 1
print("Kết quả:", ket_qua)
```
