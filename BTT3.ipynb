### W3A1
Đảo ngược số có 3 chữ số (không dùng chuỗi)

```python
n = int(input())
if 100 <= n <= 999:
    t = n // 100
    c = (n // 10) % 10
    dv = n % 10

    dao = dv * 100 + c * 10 + t
    print("Số đảo ngược:", dao)
```

### W3A3
Kiểm tra số có phải lũy thừa của 2

```python
n = int(input())
if n > 0 and (n & (n - 1)) == 0:
    print("Là lũy thừa của 2")
else:
    print("Không phải lũy thừa của 2")
```

### W3A4
Chia và làm tròn xuống

```python
m = int(input("Nhập m: "))
n = int(input("Nhập n: "))
thuong = m / n
if thuong >= 0:
    ket_qua = int(thuong)
else:
    ket_qua = int(thuong) - 1

print("Kết quả làm tròn xuống:", ket_qua)
```

### W3A17
Cấp số nhân

```python
a = int(input("Nhập số a: "))
b = int(input("Nhập số b: "))
c = int(input("Nhập số c: "))
d = int(input("Nhập số d: "))
is_gp = False 
if a == 0:
    if b == 0 and c == 0 and d == 0:
        is_gp = True
else:
    if b % a == 0:
        q = b // a
        if (b * q == c) and (c * q == d):
            is_gp = True
if is_gp:
    print("CÓ thể tạo thành cấp số nhân với công bội nguyên.")
else:
    print("KHÔNG thể tạo thành cấp số nhân với công bội nguyên.")
```
