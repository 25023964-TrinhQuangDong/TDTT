```python
def is_armstrong(x):
    if x < 0:
        return False
    temp = x
    count = 0
    t = x
    while t > 0:
        count += 1
        t //= 10
    
    total = 0
    t = x
    while t > 0:
        digit = t % 10
        
        power = 1
        i = 0
        while i < count:
            power *= digit
            i += 1
        
        total += power
        t //= 10
    
    return total == x
```

### W5A20

```python
def detect_number_type(s):
    has_dot = False
    i = 0
    while i < len(s):
        if s[i] == '.':
            if has_dot:
                return 'neither'
            has_dot = True
        elif not ('0' <= s[i] <= '9'):
            return 'neither'
        i += 1
    
    if has_dot:
        return 'float'
    return 'int'
```

### W5A25

```python
def area(x1,y1,x2,y2,x3,y3):
    a = x1*(y2-y3) + x2*(y3-y1) + x3*(y1-y2)
    if a < 0:
        a = -a
    return a

def point_in_triangle(x1,y1,x2,y2,x3,y3,px,py):
    A = area(x1,y1,x2,y2,x3,y3)
    A1 = area(px,py,x2,y2,x3,y3)
    A2 = area(x1,y1,px,py,x3,y3)
    A3 = area(x1,y1,x2,y2,px,py)
    
    return A == A1 + A2 + A3

def point_in_quadrilateral(quad, p):
    x1,y1 = quad[0]
    x2,y2 = quad[1]
    x3,y3 = quad[2]
    x4,y4 = quad[3]
    px,py = p
    
    return (point_in_triangle(x1,y1,x2,y2,x3,y3,px,py) or
            point_in_triangle(x1,y1,x3,y3,x4,y4,px,py))
```

### W5A28

```python
def sqrt(x):
    i = 0
    while i * i <= x:
        i += 1
    return i - 1  

def angle_ABC(x1, y1, x2, y2, x3, y3):
    BAx = x1 - x2
    BAy = y1 - y2
    BCx = x3 - x2
    BCy = y3 - y2

    dot = BAx * BCx + BAy * BCy

    lenBA = sqrt(BAx*BAx + BAy*BAy)
    lenBC = sqrt(BCx*BCx + BCy*BCy)

    if lenBA == 0 or lenBC == 0:
        return 0

    cos_val = dot / (lenBA * lenBC)

    angle = (1 - cos_val) * 90   # gần đúng

    return angle
```

```python
def max_min(a, b, c):
    max_v = a
    min_v = a
    
    if b > max_v:
        max_v = b
    if c > max_v:
        max_v = c
    
    if b < min_v:
        min_v = b
    if c < min_v:
        min_v = c
    
    return [max_v, min_v]
```

### W5A30

```python
def he_so_nhi_thuc(n, k):
    if k < 0 or k > n:
        return 0
    
    if k > n - k:
        k = n - k
    
    result = 1
    i = 1
    while i <= k:
        result = result * (n - k + i) // i
        i += 1
    
    return result
```

### W5A33

```python
def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    if car_type == "standard":
        base = 10000
        per_km = 12000
    else:
        base = 15000
        per_km = 15000
    
    fare = base + km * per_km
    
    fare += wait_minutes * 1000
    
    if start_hour >= 22 or start_hour < 5:
        fare = fare + fare * 0.2
    
    if km > 10:
        fare = fare - fare * 0.1
    
    return fare
```

### W5A34

```python
def has_unique_digits(x):
    seen = [0]*10
    
    while x > 0:
        d = x % 10
        if seen[d] == 1:
            return False
        seen[d] = 1
        x //= 10
    
    return True

def sum_1_to_n(n):
    if n <= 0:
        return 0
    
    total = 0
    i = 1
    while i <= n:
        if has_unique_digits(i):
            total += i
        i += 1
    
    return total
```
