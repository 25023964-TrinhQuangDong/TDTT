### W6A9

```python
d1_input = input().split()
d2_input = input().split()
d1 = {}
d2 = {}
for p in d1_input:
    k, v = p.split(':')
    d1[k] = int(v)
for p in d2_input:
    k, v = p.split(':')
    d2[k] = int(v)
res = {}
for k in d1:
    res[k] = d1[k]
for k in d2:
    if k in res:
        res[k] = res[k] + d2[k]
    else:
        res[k] = d2[k]
keys = list(res.keys())
for i in range(len(keys)):
    for j in range(i + 1, len(keys)):
        if keys[i] > keys[j]:
            tmp = keys[i]
            keys[i] = keys[j]
            keys[j] = tmp
final = {}
for k in keys:
    final[k] = res[k]
print(final)
```

### W6A16

```python
n = int(input())
m = int(input())
a = []
for i in range(n):
    row = []
    for j in range(m):
        x = int(input())
        row.append(x)
    a.append(row)
def count_digits(x):
    if x == 0:
        return 1
    cnt = 0
    if x < 0:
        cnt += 1
        x = -x
    while x > 0:
        cnt += 1
        x //= 10
    return cnt
for i in range(n):
    for j in range(m):
        d = count_digits(a[i][j])
        for k in range(4 - d):
            print(" ", end="")
        print(a[i][j], end="")
    print()
```

```python
def find_max_index(lst):
    if len(lst) == 0:
        return -1
    max_val = lst[0]
    pos = 0
    for i in range(1, len(lst)):
        if lst[i] > max_val:
            max_val = lst[i]
            pos = i
    return pos
n = int(input())
lst = []
for i in range(n):
    lst.append(int(input()))
print(find_max_index(lst))
```
