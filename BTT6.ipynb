### W6A9

```python
d1_input = input().split()
d2_input = input().split()
d1 = {}
d2 = {}
for p in d1_input:
    k, v = p.split(':')
    d1[k] = int(v)
for p in d2_input:
    k, v = p.split(':')
    d2[k] = int(v)
res = {}
for k in d1:
    res[k] = d1[k]
for k in d2:
    if k in res:
        res[k] = res[k] + d2[k]
    else:
        res[k] = d2[k]
keys = list(res.keys())
for i in range(len(keys)):
    for j in range(i + 1, len(keys)):
        if keys[i] > keys[j]:
            tmp = keys[i]
            keys[i] = keys[j]
            keys[j] = tmp
final = {}
for k in keys:
    final[k] = res[k]
print(final)
```

### W6A10

```python
nums = list(map(int, input().split()))
k = int(input())
pairs = []
for i in range(len(nums)):
    for j in range(i + 1, len(nums)):
        if nums[i] + nums[j] == k:
            pairs.append((nums[i], nums[j]))
for i in range(len(pairs)):
    for j in range(i + 1, len(pairs)):
        if pairs[i][0] > pairs[j][0]:
            tmp = pairs[i]
            pairs[i] = pairs[j]
            pairs[j] = tmp

print(pairs)
```

### W6A12

```python
nums = list(map(int, input().split()))
d = {}
for x in nums:
    if x in d:
        d[x] = d[x] + 1
    else:
        d[x] = 1
max_count = -1
result = None
for k in d:
    if d[k] > max_count:
        max_count = d[k]
        result = k
    elif d[k] == max_count:
        if k < result:
            result = k
print(result)
```

### W6A14

```python
list1 = list(map(int, input().split()))
list2 = list(map(int, input().split()))
res = []
for x in list1:
    found = False
    for y in list2:
        if x == y:
            found = True
            break
    if found:
        exists = False
        for z in res:
            if z == x:
                exists = True
                break
        if not exists:
            res.append(x)
print(res)
```

### W6A15

```python
n = int(input())
d = {}
i = 0
while i < n:
    key = input()
    value = int(input())
    d[key] = value
    i = i + 1
k = int(input())
new_d = {}
for key in d:
    if d[key] > k:
        new_d[key] = d[key]
print(new_d)
```
